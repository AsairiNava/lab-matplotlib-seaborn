#### Use the appropriate visulalization to show the distribution of the column `Age`

```python
# your code here
fig = plt.figure(figsize=[12,6])
plt.hist(x=titanic['Age'], color='#F2AB6D', rwidth=0.85)
plt.style.use('classic')
plt.title('Histograma de edades')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')

plt.show()
```

#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
fig = plt.figure(figsize=[12,6])

plt.subplot(2, 2, 1)
plt.hist(x=titanic['Age'], bins=(10), rwidth=0.85)
plt.style.use('classic')

plt.subplot(2, 2, 2)
plt.hist(x=titanic['Age'], bins=(20), rwidth=0.85)
plt.style.use('classic')

plt.subplot(2, 2, 3)
plt.hist(x=titanic['Age'], bins=(50), rwidth=0.85)
plt.style.use('classic')

plt.show()
```

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the appropriate plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.

*Hint: One of the ways involves using Seaborn*

```python
#your code-1st way
sns.lmplot(x='Fare', y='Age', data=titanic, fit_reg=False)
```

```python
# your code-2nd way
titanic.plot(x='Fare', y='Age', style='o')
plt.title('Linear relationship between Fare and Age')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.show()
```

#### Using Seaborn plot the correlation matrix of various features.

*Hint: search how to use `heatmap`*

```python
#your code here
titanic_corr = titanic[['PassengerId','Survived','Pclass','Age','SibSp', 'Parch', 'Fare']].corr()
cmap = sns.diverging_palette(6,255,as_cmap=True)
sns.heatmap(titanic_corr,cmap=cmap,annot=True)
```

#### According to the previous plot, what are the most correlated feature? Comment.

```python
#your comment here

#The relationship between Pclas and Fare is the lowest, 
#as is the relationship between Parch and Sib Sp is the highest, although without reaching 1
```
